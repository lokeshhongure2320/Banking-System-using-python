```python
class bankaccount():
    def __init__(self,acc_no,name,initial_bal,acc_type):
        self.acc_no = acc_no
        self.name = name
        self.initial_bal = initial_bal
        self.acc_type = acc_type

    def deposite(self,amount):
        if amount>0:
            self.initial_bal += amount
        print("deposite: ",amount)
        print("initial balance is: ",self.initial_bal)

    def withdraw(self,amount):
        if amount>100 and self.initial_bal > (500+amount):
            self.initial_bal -= amount
            print("withdraw succesfully")
        else:
            print("no efficient balance")
        print("current balance is: ",self.initial_bal)

    def getdetails(self):
        return {"Account_no":self.acc_no,
                "Name":self.name,
                "balance":self.initial_bal,
                "account_type":self.acc_type}
```

```python
c1 = bankaccount(123,"kunal",10000,"saving")
```

```python
c1.deposite(1000)
```

```python
c1.withdraw(1000)
```

```python
c1.getdetails()
```
